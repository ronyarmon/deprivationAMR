```python
#import packages 
import statsmodels.api as sm
from statsmodels.api import add_constant
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#read deprivation ranks and %diff, by AreaCode
directory_path=##[place your diretory path here]##
file_name='desc_IMD_92377.csv'
file_path=directory_path+file_name
depr_desc=pd.read_csv(file_path)
del depr_desc['Unnamed: 0']

#lists and constants
indices_list=depr_desc.columns[5:] #list of variables
emp_edu_he=['Employment','Education','Health'] 
models_dict={}
slopes_list=[]
rsquared_list=[]
p_values_list=[]
signif_codes=[(0,'***'), (0.001,'***'),(0.01,'**'),(0.05,'*'),(0.1,'.'),(1,'')]
sl_len=len(signif_codes)-1
y=depr_desc["mean_presc"]

for index, value in enumerate (indices_list):
    #run models for deprivation indices
    X=add_constant(depr_desc[value])
    model = sm.OLS(y,X).fit()
    models_dict[value]=model
    
    #get slope 
    slope=round(model.params[1],5)
    slopes_list=slopes_list+[slope]

    #get Rsquared values for the slopes
    r2=round(model.rsquared,5)
    rsquared_list=rsquared_list+[r2]
    
    #get p_values and significant coding
    p_value=round(model.pvalues[1],4)
    for index,level in enumerate (signif_codes):
        if index<sl_len:
            low,high,code=signif_codes[index][0],signif_codes[index+1][0],signif_codes[index+1][1]
            if low <= p_value < high:
                p_code='<'+str(high)+' '+code
                p_values_list=p_values_list+[p_code]
```

```python
#Employment + Education + Health 
y=depr_desc["mean_presc"]
X_he_ed=add_constant(depr_desc[['Health','Education']])
X_he_emp=add_constant(depr_desc[['Health','Employment']])
X_ed_emp=add_constant(depr_desc[['Education','Employment']])
X_he_ed_emp=add_constant(depr_desc[['Health','Education','Employment']])

model_he_ed = sm.OLS(y,X_he_ed).fit()
model_he_emp = sm.OLS(y,X_he_emp).fit()
model_ed_emp = sm.OLS(y,X_ed_emp).fit()
model_he_ed_emp=sm.OLS(y,X_he_ed_emp).fit()

print ('\n\nHealth & Education\n----------------------------\n\n',model_he_ed.summary())
print ('\nHealth & Employment\n----------------------------\n\n',model_he_emp.summary())
print ('\nEducation & Employment\n----------------------------\n\n',model_ed_emp.summary())
print ('\nHealth, Education & Employment\n----------------------------\n\n',model_he_ed_emp.summary())
```
