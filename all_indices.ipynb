```python
#import packages 
import statsmodels.api as sm
from statsmodels.api import add_constant
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#read deprivation ranks and %diff, by AreaCode
directory_path=##[place your diretory path here]##
file_name='desc_IMD_92377.csv'
file_path=directory_path+file_name
depr_desc=pd.read_csv(file_path)

indices_list=depr_desc.columns[7:] #list of variables
models_dict={}
slopes_list=[]
rsquared_list=[]
p_values_list=[]
signif_codes=[(0,'***'), (0.001,'***'),(0.01,'**'),(0.05,'*'),(0.1,'.'),(1,'')]
sl_len=len(signif_codes)-1
y=depr_desc["mean_presc"]

for index, value in enumerate (indices_list):
    #run model for deprivation index
    X=add_constant(depr_desc[value])
    model = sm.OLS(y,X).fit()
    models_dict[value]=model
    
    #get slope 
    slope=round(model.params[1],5)
    slopes_list=slopes_list+[slope]

    #get Rsquared values for the slopes
    r2=round(model.rsquared,5)
    rsquared_list=rsquared_list+[r2]
    
    #get p_values and significant coding
    p_value=round(model.pvalues[1],4)
    for index,level in enumerate (signif_codes):
        if index<sl_len:
            low,high,code=signif_codes[index][0],signif_codes[index+1][0],signif_codes[index+1][1]
            if low <= p_value < high:
                p_code='<'+str(high)+code
                if p_code=='<1':
                    p_code='>0.1   '
                p_values_list=p_values_list+[p_code]

#print paraemeters for all models
models_summary=pd.DataFrame()
models_summary['Slope']=slopes_list
models_summary['p value']=p_values_list
models_summary['Rsquared']=rsquared_list
models_summary=models_summary.set_index(indices_list)
print ('\n\nSummary')
print ('=======')
print (models_summary)
models_summary.to_csv('models_summary.csv')


#print and chart all models
print ('\n\n\nIndices: Models and Charts')
print ('==================')

for key, model in models_dict.items():
    print ('\n\n',key,'\n',model.summary(),'\n')
    sns.lmplot(x= key,y= "mean_presc", data=depr_desc) #plot regression line
    plt.ylabel('Mean Prescribing')
    plt.show()
```
